使用循环和列表推导找出单词长度大于某个数字的单词<br>
words = ["apple", "banana", "orange", "peach","kiwi"]<br>

```python
words=['apple','banana','orange','peach','kiwi']
def findlen(n):
    bigthan=[]
    for i in words:
        if int(len(i))>n:
            bigthan.append(i)
    return bigthan
findlen(5)
```

```python
words=['apple','banana','orange','peach','kiwi']
[i for i in words if int(len(i))>5]
```

寻找两个列表中的相同元素<br>
l1=[1,2,5]<br>
l2=[6,2,7]<br>

```python
l1=[1,2,5];l2=[6,2,7]
tup=[(i,j) for i in l1 for j in l2]
[k for k in tup if k[0]==k[1]]
```

实现求指定长度的Fibonacci 数列<br>
Fib数组初始为[0,1]<br>
分别要求使用循环和数组实现<br>

```python
def fibonacci(n):
    Fib=[0,1]
    a=Fib[0]
    b=Fib[1]
    i=2
    while i<n:
        tmp=a
        a=b
        b=a+tmp
        Fib.append(b)
        i+=1
    return Fib
#test
fibonacci(10)        
```

```python
import numpy as np
def fibonacci2(n):
    Fib=[0,1]
    a=np.zeros(n)
    a[0]=Fib[0];a[1]=Fib[1]
    for i in range(2,n):
        a[i]=a[i-1]+a[i-2]
        Fib.append(int(a[i]))
    return Fib
#test
fibonacci2(10)
```

[ ]对应检查<br> 
输入含有[]的字符串，输出对中括号出现规则的检测结果<br>
 []        OK   ][        NOT OK<br>

 [][]      OK   ][][      NOT OK<br>

 [[][]]    OK   []][[]    NOT OK<br>

 [[][[]]]  OK   ][]][[][  NOT OK<br>

```python
s=input('type several [ and ]:')
dictbase={'[':0,']':0}
if s.count('[')!=s.count(']'):
    print("'[' and ']'have different number")
else:
    for i in s:
        if i=='[':
            dictbase['[']+=1
        elif (i==']') & (dictbase[']']<dictbase['[']):
            dictbase[']']+=1
        else:
            pass

if (dictbase['[']==dictbase[']']) & (dictbase['[']!=0):
    print("OK!")
elif dictbase['[']!=0:
    print("NOT OK!")
else:
    pass
```
